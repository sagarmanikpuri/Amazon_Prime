```python
#Import Important Libraries
from google.colab import files
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Upload xlsx Data
from google.colab import files

#upload the excel file
uploaded = files.upload() #This will open a file picker in colab

#Read each uploaded files into pandas dataframe
dataframes = {} #Dictionary to store dataframes

for filename in uploaded.keys():
  df = pd.read_excel(filename)
  dataframes[filename] = df
  print(f"✅ '{filename}' uploaded successfully! Shape: {df.shape}")

# Step 4: Access each DataFrame by filename
# Example:
# df1 = dataframes['file1.xlsx']
# df2 = dataframes['file2.xlsx']

# Step 5 (Optional): Display first few rows of both DataFrames
for name, df in dataframes.items():
    print(f"\n📘 Preview of {name}:")
    print(df.head())
```

```python
df_credit = pd.read_excel("credits.xlsx")
df_title = pd.read_excel("title.xlsx")
```

```python
# 🧹 Data Cleaning for both DataFrames

# For df_credit
print("🧾 Cleaning df_credit ...")
df_credit_before = df_credit.shape[0]

# 1️⃣ Remove duplicate rows
df_credit = df_credit.drop_duplicates()
print(f"✅ Removed {df_credit_before - df_credit.shape[0]} duplicate rows from df_credit")

# 2️⃣ Replace NaN / missing values
# For numeric columns → replace NaN with 0
# For object (text) columns → replace NaN with "Unknown"
df_credit = df_credit.fillna({
    col: 0 if pd.api.types.is_numeric_dtype(df_credit[col]) else "Unknown"
    for col in df_credit.columns
})

print("✅ Replaced all NaN values in df_credit")

# ---------------------------------------------------------------------

# For df_title
print("\n🎬 Cleaning df_title ...")
df_title_before = df_title.shape[0]

# 1️⃣ Remove duplicate rows
df_title = df_title.drop_duplicates()
print(f"✅ Removed {df_title_before - df_title.shape[0]} duplicate rows from df_title")

# 2️⃣ Replace NaN / missing values
df_title = df_title.fillna({
    col: 0 if pd.api.types.is_numeric_dtype(df_title[col]) else "Unknown"
    for col in df_title.columns
})

print("✅ Replaced all NaN values in df_title")

# ---------------------------------------------------------------------

# Optional: Verify cleaning results
print("\n📊 After Cleaning:")
print(f"df_credit shape: {df_credit.shape}")
print(f"df_title shape: {df_title.shape}")

# Check if any NaN remains
print("\n🔎 Remaining missing values:")
print("df_credit:", df_credit.isna().sum().sum())
print("df_title:", df_title.isna().sum().sum())
```

```python
# 🔍 Check Data Types and Null Counts

print("📘 df_credit Info:")
print("-" * 50)
print(df_credit.info())

print("\n🎬 df_title Info:")
print("-" * 50)
print(df_title.info())
```

```python
# 📊 Describe Data (summary statistics for all columns)
print("📘 df_credit Summary:")
print("-" * 60)
print(df_credit.describe(include="all"))

print("\n🎬 df_title Summary:")
print("-" * 60)
print(df_title.describe(include="all"))
```

```python
#Make all column name uppercase
df_title.columns = df_title.columns.str.upper()

#verify
print(df_title.columns)
```

```python
#Merge Data
df = pd.merge(df_credit, df_title, on="ID", how="inner")
```

```python
# 📊 Basic Structure of the merged DataFrame

# 1️⃣ Rows and Columns
print("🔹 Shape of DataFrame (rows, columns):", df.shape)

# 2️⃣ Column Names
print("\n🔹 Column Names:")
print(df.columns)

# 3️⃣ Data Types and Non-Null Counts
print("\n🔹 DataFrame Info:")
df.info()

# 4️⃣ Summary Statistics (numeric + categorical)
print("\n🔹 Summary Statistics:")
df.describe(include='all')
```

```python
# Set style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

# -----------------------------
# 1️⃣ Count of content types
# -----------------------------
plt.figure()
sns.countplot(x='TYPE', data=df, palette='Set2')
plt.title("Count of Content by Type (Movie / Series)")
plt.show()

# -----------------------------
# 2️⃣ Distribution of release years
# -----------------------------
plt.figure()
sns.histplot(df['RELEASE_YEAR'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
# 3️⃣ Genre frequency (Top 20)
# -----------------------------
# Split multiple genres and count
from collections import Counter
all_genres = df['GENRES'].dropna().apply(lambda x: str(x).split(','))
flat_genres = [genre.strip() for sublist in all_genres for genre in sublist]
top_genres = dict(Counter(flat_genres).most_common(20))

plt.figure()
sns.barplot(x=list(top_genres.values()), y=list(top_genres.keys()), palette='magma')
plt.title("Top 20 Genres")
plt.xlabel("Count")
plt.show()

# -----------------------------
# 4️⃣ Runtime distribution
# -----------------------------
plt.figure()
sns.histplot(df['RUNTIME'], bins=30, kde=True, color='salmon')
plt.title("Runtime Distribution")
plt.xlabel("Runtime (minutes)")
plt.show()
```

```python
# 5️⃣ IMDB and TMDB Scores
# -----------------------------
plt.figure()
sns.histplot(df['IMDB_SCORE'], bins=20, kde=True, color='green')
plt.title("IMDB Score Distribution")
plt.xlabel("IMDB Score")
plt.show()

plt.figure()
sns.histplot(df['TMDB_SCORE'], bins=20, kde=True, color='purple')
plt.title("TMDB Score Distribution")
plt.xlabel("TMDB Score")
plt.show()

# -----------------------------
# 6️⃣ Age Certification counts
# -----------------------------
plt.figure()
sns.countplot(y='AGE_CERTIFICATION', data=df, order=df['AGE_CERTIFICATION'].value_counts().index)
plt.title("Count by Age Certification")
plt.show()
```

```python
# 7️⃣ Seasons distribution for series
# -----------------------------
plt.figure()
sns.histplot(df[df['TYPE']=='TV Show']['SEASONS'], bins=20, color='orange', kde=False)
plt.title("Seasons Distribution for TV Shows")
plt.xlabel("Number of Seasons")
plt.show()

# -----------------------------
# 8️⃣ Popularity vs IMDB Score
# -----------------------------
plt.figure()
sns.scatterplot(x='TMDB_POPULARITY', y='IMDB_SCORE', hue='TYPE', data=df, palette='Set1')
plt.title("TMDB Popularity vs IMDB Score")
plt.show()
```
